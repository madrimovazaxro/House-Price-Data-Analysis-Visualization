```python
cat_all = df.select_dtypes(include = 'object').columns
df[cat_all] = df[cat_all].fillna('None')

num = df.select_dtypes(include = 'number').columns
df[num] = df[num].fillna(df[num].median())
```

```python
df.isnull().sum().sum()
```

```python
sns.set_theme()
```

```python
#1. Histogram
sns.histplot(df['SalePrice'], color = '#2ECB9b', kde = True, bins = 10)
plt.show()
```

```python
#2.KDE
sns.kdeplot(df['SalePrice'], bw_adjust = 0.1)
plt.show()
```

```python
#3. Boxplot
sns.boxplot(data = df, y = 'SalePrice')
plt.show()
```

```python
fig,axes = plt.subplots(1, 3, figsize = (16,8))
#1.Histogram
sns.histplot(df['SalePrice'], ax = axes[0], color = '#2ECB9b', kde = True, bins = 10)
axes[0].set_title("Taqsimot")
axes[0].tick_params(axis = 'x', rotation=45)
#2.KDE
sns.kdeplot(df['SalePrice'], ax = axes[1], fill=True, color = "#2ECB9b", bw_adjust=0.2)
axes[1].set_title('KDE egri chiziq\'i')
axes[1].tick_params(axis = 'x', rotation =45)
#3. Boxplot
sns.boxplot(y=df['SalePrice'], ax = axes[2], color= "#2ECB9b")
axes[2].set_title('Boxplot-outlier')
plt.tight_layout()
plt.show()
```

```python
df['SalePrice_log'] = np.log1p(df['SalePrice'])
fig,axes = plt.subplots(1,2, figsize=(16,8))

sns.histplot(df['SalePrice'], ax=axes[0], color='#EF4444', kde=True)
axes[0].set_title('Asl')

sns.histplot(df['SalePrice_log'], ax=axes[1], color= '#EF4444', kde=True)
axes[1].set_title("Log - normal")

plt.tight_layout()
plt.show()
```

```python
#Raqamli ustunlar
num_df = df.select_dtypes(include ="number")
corr = num_df.corr()

#SalePrice bilan eng kuchli correlationga ega top 10 featurelarni topadi.
top10 = corr["SalePrice"].drop("SalePrice").abs().sort_values(ascending=False).head(30).index

#O‘sha featurelarga SalePrice ni qo‘shib yangi list yaratadi.
top_cols = list(top10) + ['SalePrice']

#Heatmap
plt.figure(figsize=(30,15))
sns.heatmap(corr[top_cols].loc[top_cols], annot=True, fmt='.2f',
            cmap='RdYlGn', center= 0, vmin=-1, vmax =1)
plt.title('Top 10 Korrelatsiya')
plt.show()
```

```python
fig,axes = plt.subplots(1, 2, figsize=(16,8))

#1.Boxplot - har sifat darajasi
sns.boxplot(data=df, x="OverallQual", y="SalePrice", palette='magma', ax=axes[0])
axes[0].set_title('Sifat vs Narx - Boxplot')

#2.Violinplot - taqsimot shakli
sns.violinplot(data=df, x="OverallQual", y="SalePrice", palette='magma', ax=axes[1])
axes[1].set_title('Sifat vs Narx - Violin')

plt.tight_layout()
plt.show()
```

```python
hood = df.groupby('Neighborhood')['SalePrice'].mean().sort_values(ascending=False)

#print(hood.index)

plt.figure(figsize = (18,6))

sns.barplot(data=df, x='Neighborhood', y="SalePrice", order=hood.index, color= '#22cce3')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize = (12,6))
#1. Oddiy scatterplot
sns.scatterplot(
    data= df,
    x='GrLivArea',
    y='SalePrice',
    hue='OverallQual',
    palette = 'RdYlGn',
    ax= axes[0]
    )
axes[0].set_title('GrLivArea va SalePrice')

#2. Regplot - trend chiziqi
sns.regplot(
    data= df,
    x='GrLivArea',
    y='SalePrice',
    scatter_kws = {'alpha':0.3, 'color':'#2ECB9B', 'edgecolor':'black'},
    line_kws={'color':'red'}, ax=axes[1]
)
axes[1].set_title("Trend")

plt.tight_layout()
plt.show()
```
